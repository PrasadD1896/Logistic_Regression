```python
dftr['Married'] = dftr['Married'].fillna(dftr['Married'].mode)
dftr['Dependents'] = dftr['Dependents'].fillna(dftr['Dependents'].mode)
dftr['Self_Employed'] = dftr['Self_Employed'].fillna(dftr['Self_Employed'].mode)
dftr['Loan_Amount_Term'] = dftr['Loan_Amount_Term'].fillna(dftr['Loan_Amount_Term'].mode)
```

```python
dfte['Gender'] = dfte['Gender'].fillna(dfte['Gender'].mode)
dfte['Dependents'] = dfte['Dependents'].fillna(dfte['Dependents'].mode)
dfte['Self_Employed'] = dfte['Self_Employed'].fillna(dfte['Self_Employed'].mode)
dfte['LoanAmount'] = dfte['LoanAmount'].fillna(dfte['LoanAmount'].median)
dfte['Loan_Amount_Term'] = dfte['Loan_Amount_Term'].fillna(dfte['Loan_Amount_Term'].median)
dfte['Credit_History'] = dfte['Credit_History'].fillna(dfte['Credit_History'].median)
```

```python
dfte = dfte.drop('Loan_ID',axis = 1)
x = dftr.drop('Loan_Status',axis = 1)
y = dftr['Loan_Status']
```

```python
dftr = pd.get_dummies(dftr)
dfte = pd.get_dummies(dfte)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_v,y_train,y_v = train_test_split(x,y,test_size = 0.3,random_state = 30)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
model = LogisticRegression()
model.fit(x_train,y_train)
```
